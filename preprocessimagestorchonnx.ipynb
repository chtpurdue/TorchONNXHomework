```python
TRAIN_DIR = "/content/drive/MyDrive/data/train"
TEST_DIR = "/content/drive/MyDrive/data/test"

SAVE_PATH_TRAIN = "/content/drive/MyDrive/data/train_data.pt"
SAVE_PATH_TEST = "/content/drive/MyDrive/data/test_data.pt"
```

```python
from PIL import Image
print("Pillow installed successfully")
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, TensorDataset
```

```python
IMAGE_SIZE = (224, 224)
```

```python
transform = transforms.Compose([
    transforms.Resize(IMAGE_SIZE), transforms.ToTensor(), transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])
```

```python
train_dataset = datasets.ImageFolder(
    root=TRAIN_DIR,
    transform=transform
)

test_dataset = datasets.ImageFolder(
    root=TEST_DIR,
    transform=transform
)
```

```python
def dataset_to_tensors(dataset):
    images = []
    labels = []

    for img, label in dataset:
        images.append(img)
        labels.append(label)

    return torch.stack(images), torch.tensor(labels)

print("Processing training data...")
train_images, train_labels = dataset_to_tensors(train_dataset)

print("Processing test data...")
test_images, test_labels = dataset_to_tensors(test_dataset)
```

```python
torch.save((train_images, train_labels), SAVE_PATH_TRAIN)
torch.save((test_images, test_labels), SAVE_PATH_TEST)

print("Saved processed data to Google Drive.")
```
